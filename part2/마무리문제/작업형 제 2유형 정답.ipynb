```python
import pandas as pd
import numpy as np
```

```python
X_train = pd.read_csv('../yemoonsaBigdata/datasets/Part2/stellar_X_train.csv')
y_train = pd.read_csv('../yemoonsaBigdata/datasets/Part2/stellar_y_train.csv')
X_test = pd.read_csv('../yemoonsaBigdata/datasets/Part2/stellar_X_test.csv')
```

```python
X_train.info()
```

```python
X_train.describe()
```

```python
'''

이상치 제거

'''

train_dataset = pd.concat([X_train, y_train], axis=1)
train_dataset = train_dataset[train_dataset['u'] != -9999]
```

```python
train_dataset.reset_index(drop=True, inplace=True)
```

```python
train_dataset.describe()
```

```python
X_train = pd.DataFrame(train_dataset.iloc[:, :-1])
y_train = pd.DataFrame(train_dataset.iloc[:, -1])
```

```python
# redshift가 이상함
X_train['redshift'].quantile([q/100 for q in range(90, 101)])
```

약 2%만이 3을 넘는다는 것을 확인할 수 있다...

<br>

```python
# 파생 변수 추가

X_train['redshift_upper_3'] = np.where(X_train['redshift'] > 3, 1, 0)
X_test['redshift_upper_3'] = np.where(X_test['redshift'] > 3, 1, 0)
```

```python
X_train.info()
```

```python
columns_remove = []
columns_num = ['alpha', 'delta', 'u', 'g', 'r', 'i', 'z', 'redshift']
columns_cat = ['run_ID', 'rerun_ID', 'cam_col', 'field_ID', 'plate', 'redshift_upper_3']
columns_y = ['galaxy']
```

```python
X_train[columns_cat] = X_train[columns_cat].astype(str)
X_test[columns_cat] = X_test[columns_cat].astype(str)
```

```python
train_dataset = pd.concat([X_train, y_train], axis=1)

for col in columns_num:
    print("-" * 80)
    print(col)
    print(train_dataset.groupby(columns_y)[col].describe(), end='\n\n')
```

수치형 중 의미있는 변수로 보이는 것은 `redshift`이다.

<br><br>

```python
for col in columns_cat:
    print(train_dataset.groupby(col)[columns_y].mean().sort_values(by=columns_y, ascending=False), end='\n\n')
```

범주형 중 의미있어 보이는 변수는 `cam_col`, `redshift_upper_3`이다.

<br>

```python
# 범주형 데이터는 Label Encoder나 OneHot Encoder로 encoding해야 한다.

from sklearn.preprocessing import LabelEncoder


X = pd.concat([X_train, X_test])

for col in columns_cat:
    label_encoder = LabelEncoder()
    label_encoder.fit(X[col])
    
    X_train[col] = label_encoder.transform(X_train[col])
    X_test[col] = label_encoder.transform(X_test[col])
```

```python
# 학습 데이터와 검증 데이터로 분리

from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.3, stratify=y_train)
```

```python
# 수치형 데이터에 대해 Scaling 진행

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_tr[columns_num])

X_tr[columns_num] = scaler.transform(X_tr[columns_num])
X_val[columns_num] = scaler.transform(X_val[columns_num])
X_test[columns_num] = scaler.transform(X_test[columns_num])
```

```python
# 학습 진행 (RandomForestClassifier)

from sklearn.ensemble import RandomForestClassifier

model_RF = RandomForestClassifier()
model_RF.fit(X_tr, y_tr.values.squeeze())
```

```python
# 학습 진행 (XGBClassifier)

from xgboost import XGBClassifier

model_XGB = XGBClassifier()
model_XGB.fit(X_tr, y_tr.values.squeeze())
```

```python
# 평가

from sklearn.metrics import roc_auc_score

y_val_pred_RF = model_RF.predict_proba(X_val)
y_val_pred_XGB = model_XGB.predict_proba(X_val)

score_RF = roc_auc_score(y_val, y_val_pred_RF[:, 1])
score_XGB = roc_auc_score(y_val, y_val_pred_XGB[:, 1])

print(score_RF)
print(score_XGB)
```
