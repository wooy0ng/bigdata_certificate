### 작업형 제2유형

```python
import numpy as np
import pandas as pd

X_train = pd.read_csv('../../yemoonsaBigdata/datasets/Part3/404_x_train.csv')
X_test = pd.read_csv('../../yemoonsaBigdata/datasets/Part3/404_X_test.csv')
y_train = pd.read_csv('../../yemoonsaBigdata/datasets/Part3/404_y_train.csv')
```

```python
X_train.info()
```

```python
X_train.describe()
```

```python
col_del = ['ID']
col_num = ['Age', 'Work_Experience', 'Family_Size']
col_cat = ['Gender', 'Ever_Married', 'Graduated', 'Profession', 'Spending_Score']
col_y = ['Segmentation']
```

```python
from sklearn.preprocessing import LabelEncoder

X = pd.concat([X_train[col_cat], X_test[col_cat]])

for col in col_cat:
    label_encoder = LabelEncoder()
    label_encoder.fit(X[col])
    
    X_train[col] = label_encoder.transform(X_train[col])
    X_test[col] = label_encoder.transform(X_test[col])
    
for col in col_y:
    y_label_encoder = LabelEncoder()
    y_label_encoder.fit(y_train[col])
    y_train[col] = y_label_encoder.transform(y_train[col])
```

```python
from sklearn.model_selection import train_test_split


X_tr, X_val, y_tr, y_val = train_test_split(X_train[col_num+col_cat], y_train[col_y], test_size=0.3, stratify=y_train[col_y])

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_tr[col_num])

X_tr[col_num] = scaler.transform(X_tr[col_num])
X_val[col_num] = scaler.transform(X_val[col_num])
X_test[col_num] = scaler.transform(X_test[col_num])
```

```python
from xgboost import XGBClassifier

model_xgb1 = XGBClassifier()
model_xgb1.fit(X_tr, y_tr)

model_xgb2 = XGBClassifier(n_estimators=50, max_depth=5)
model_xgb2.fit(X_tr, y_tr)

model_xgb3 = XGBClassifier(n_estimators=50, max_depth=10)
model_xgb3.fit(X_tr, y_tr)

model_xgb4 = XGBClassifier(n_estimators=100, max_depth=5)
model_xgb4.fit(X_tr, y_tr)

model_xgb5 = XGBClassifier(n_estimators=100, max_depth=10)
model_xgb5.fit(X_tr, y_tr)
```

```python
from sklearn.metrics import f1_score

y_val_pred_1 = model_xgb1.predict(X_val)
score = f1_score(y_val, y_val_pred_1, average='macro')
print(score)

y_val_pred_2 = model_xgb2.predict(X_val)
score = f1_score(y_val, y_val_pred_2, average='macro')
print(score)

y_val_pred_3 = model_xgb3.predict(X_val)
score = f1_score(y_val, y_val_pred_3, average='macro')
print(score)

y_val_pred_4 = model_xgb4.predict(X_val)
score = f1_score(y_val, y_val_pred_4, average='macro')
print(score)

y_val_pred_5 = model_xgb5.predict(X_val)
score = f1_score(y_val, y_val_pred_5, average='macro')
print(score)
```

```python
pred = model_xgb2.predict(X_test[col_num+col_cat])
pred = y_label_encoder.inverse_transform(pred)

result = pd.DataFrame({
    'ID': X_test['ID'],
    'Segmentation': pred
})

result.to_csv('../../yemoonsaBigdata/res/003000000.csv', index=False)
```
